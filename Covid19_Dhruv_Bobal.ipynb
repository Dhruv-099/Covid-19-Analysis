```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot  as plt
```

```python
dataset=pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
#delete useless columns
df= dataset.drop(["Lat","Long" ], axis =1, inplace =True)
```

```python
#aggregate the rows by the country
corona_df_aggregate= dataset.groupby("Country/Region").sum()
```

```python
corona_df_aggregate.head()
```

```python
corona_df_aggregate.shape
```

```python
#Visualize the data related to Country
corona_df_aggregate.loc["India"].plot()
corona_df_aggregate.loc["Spain"].plot()
corona_df_aggregate.loc["Russia"].plot()
plt.legend()
```

```python
#Calculate a good measure
corona_df_aggregate.loc['India'].plot()
```

```python
corona_df_aggregate.loc['India'][:120].plot()
```

```python
corona_df_aggregate.loc['India'].diff().plot()
```

```python
#Maximum infection rate for India
corona_df_aggregate.loc['India'].diff().max()
```

```python
corona_df_aggregate.loc['Spain'].diff().max()
```

```python
countries= list(corona_df_aggregate.index)
max_infection_rate=[]
for c in countries:
    max_infection_rate.append(corona_df_aggregate.loc[c].diff().max())
corona_df_aggregate["max_infection_rate"]= max_infection_rate
```

```python
corona_df_aggregate.head()
```

```python
corona_data= pd.DataFrame(corona_df_aggregate["max_infection_rate"])
corona_data
```

```python
#importing happiness report
happiness_report= pd.read_csv("worldwide_happiness_report.csv")
happiness_report.head()
```

```python
not_req_cols=["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happiness_report.drop(not_req_cols, axis=1, inplace=True)
```

```python
happiness_report.set_index("Country or region", inplace=True)
happiness_report.head()
```

```python
corona_data.shape
```

```python
happiness_report.shape
```

```python
data= corona_data.join(happiness_report, how="inner")
data
```
